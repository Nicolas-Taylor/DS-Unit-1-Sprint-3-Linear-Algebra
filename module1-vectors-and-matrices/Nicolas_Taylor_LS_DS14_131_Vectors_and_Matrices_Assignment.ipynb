## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}

\end{bmatrix}
\qquad
J= 
\begin{bmatrix}

```
#Creates new matrix for "H" and "J"

H = np.array([[12,11],
              [7,10]])
J = np.array([[0,1,2],
              [7,10,4],
              [3,2,0]])

#Finds determinant of "H" and "J"
print('Determinant of H is:', np.linalg.det(H))
print('Determinant of J is:', np.linalg.det(J))
```
